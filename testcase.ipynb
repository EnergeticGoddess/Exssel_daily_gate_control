```python
import pandas as pd
```

```python
# Читаем данные
df1 = pd.read_excel('2024-08-06.xls')
df1.head(20)
```

```python
#Формируем таблицу и столбцы
df1 = df1.drop(index=[0,1,2,3,4,5,6])
df1 = df1.rename(columns=
                {'Журнал событий входа-выхода.': 'Date',
                 'Unnamed: 1': 'Department',
                 'Unnamed: 2': 'Pass№', 'Unnamed: 3': 'Position',
                 'Unnamed: 4': 'Time in', 'Unnamed: 5': 'Time out'})
df1 = df1.reset_index(drop=True)
df1.head(10)
```

```python
# Приводим пропуски к общему вдиу
df1['Time in'] = df1['Time in'].replace('(нет)', pd.NA)
df1['Time out'] = df1['Time out'].replace('(нет)', pd.NA)
Nan = df1['Date'][1]
df1 = df1.replace(Nan, pd.NA)
df1.head(10)
```

```python
# Обрабатываем пустые строки посещений, добавляем уникальный идентификатор № посетителя для каждой строки.
# Таб№ столбец в исходных данных с этим не справляется из-за периодического дублирования названия компании без доп. идентификации, хотя на уровне данных она явно есть.
df1['Visit ID'] = None
visit_id = 0
for i in range(len(df1)):
    if pd.notna(df1.loc[i, 'Date']) and df1.loc[i, 'Date'] != '<NA>':
        visit_id += 1
    df1.loc[i, 'Visit ID'] = visit_id

unique_visits = df1['Visit ID'].unique()
for visit in unique_visits:
    visit_group = df1[df1['Visit ID'] == visit]
    filled_position = visit_group['Position'].first_valid_index()
    if filled_position is not None:
        position_value = df1.loc[filled_position, 'Position']
        df1.loc[df1['Visit ID'] == visit, 'Position'] = df1.loc[
            df1['Visit ID'] == visit, 'Position'].fillna(position_value)
df1[['Department', 'Date', 'Pass№']] = df1[
    ['Department', 'Date', 'Pass№']].fillna(method='ffill')
df1.head(10)
```

```python
# Разделяем строчки по признаку Выход\Вход
new_rows = []
for index, row in df1.iterrows():
    if pd.notna(row['Time in']):
        new_row = row.copy()
        new_row['Time'] = row['Time in']
        new_row['In/Out'] = 'In'
        new_rows.append(new_row)
    
    if pd.notna(row['Time out']):
        new_row = row.copy()
        new_row['Time'] = row['Time out']
        new_row['In/Out'] = 'Out'
        new_rows.append(new_row)

df1 = pd.DataFrame(new_rows)
df1 = df1.drop(columns=['Time in', 'Time out'])
df1 = df1.reset_index(drop=True)
df1.head(10)
```

```python
# Создаем столбец прохода через конкретное КПП
df1['Gate'] = df1['Time'].str.extract(r'\((\d+)\)')
df1['Time'] = df1['Time'].str.replace(r'\s*\(\d+\)', '', regex=True)
df1.head(10)
```

```python
# Даем таблице имя стобца и убираем не несущий смысла столбец
df1.name = '2024-08-06, Таблица контроля посещения персоналом объекта Х'
df1= df1.drop(columns=['Date'])
df1.head(10)
```

```python
# Убираем лишние пробелы для правильной группировки данных
df1 = df1.applymap(lambda x: x.replace(' ', '') if isinstance(x, str) else x)
df1.head(10)
```
